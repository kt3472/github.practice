LCEL 인터페이스
  - 사용자 정의 체인을 가능한 쉽게 만들 수 있도록, Runnable 프로토콜을 구현했습니다.

  - Runnable 프로토콜은 대부분의 컴포넌트에 구현되어 있습니다.

  - 이는 표준 인터페이스로, 사용자 정의 체인을 정의하고 표준 방식으로 호출하는 것을 쉽게 만듭니다. 표준 인터페이스에는 다음이 포함됩니다.

    1. stream: 응답의 청크를 스트리밍합니다.
    2. invoke: 입력에 대해 체인을 호출합니다.
    3. batch: 입력 목록에 대해 체인을 호출합니다.

  - 비동기 메소드도 있습니다.

    1. astream: 비동기적으로 응답의 청크를 스트리밍합니다.
    2. ainvoke: 비동기적으로 입력에 대해 체인을 호출합니다.
    3. abatch: 비동기적으로 입력 목록에 대해 체인을 호출합니다.
    4. astream_log: 최종 응답뿐만 아니라 발생하는 중간 단계를 스트리밍합니다.

체인을 실행하는 방법
  - LangChain의 'Runnable' 프로토콜
    1. invoke : 주어진 입력에 대해 체인을 호출하고, 결과를 반환한다. 이 메소드는 단일 입력에 대해 동기적으로 작동한다
    2. batch : 입력 리스트에 대해 체인을 호출하고, 각 입력에 대한 결과를 리스트로 반환한다. 이 메소드는 여러 입력에 대해 동기적으로 작동하며, 표율적인 배치처리를 한다
    3. stream : 입력에 대해 체인을 호출하고, 결과의 조각들을 스트리밍한다. 대용량 처리나 실시간 데이터 처리에 유용하다
    4. 비동기버전 : ainvoke, abatch, astream 등의 메소드는 각각의 비동기 실행을 지원한다.

LangChain을 사용하여 커스텀 체인을 생성하는 과정
  1. 필요한 컴포넌트를 정의하고, 각각 'Runnalbe' 인터페이스 구현
  2. 컴포넌트들을 조합하여 사용자 정의 체인 생성
  3. 생성된 체인을 사용하여 데이터 처리 작업을 수행한다.

```python
# 필요한 라이브러리 임포트
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import asyncio
import os

# OpenAI API 키 환경 변수에 설정 (본인 키로 변경)
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")


# 모델 인스턴스 생성 (API 키를 따로 전달 안 하면 환경변수에서 자동 읽음)
model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

# 프롬프트 템플릿 생성: "{topic} 에 대하여 3문장으로 설명해줘."
prompt = PromptTemplate.from_template("{topic} 에 대하여 3문장으로 설명해줘.")

# 프롬프트와 모델, 출력 파서 연결하여 체인 생성
chain = prompt | model | StrOutputParser()

# 동기 스트림 실행 예시: '멀티모달' 토픽에 대한 토큰 스트림 출력
print("== 동기 스트림 출력 ==")
# .invoke() → 결과를 한 번에 받음.
# .stream() → 토큰이 생기는 즉시 실시간으로 받아옴.
# for token in chain.stream(...) → 한 글자씩 찍히는 것처럼 나옴.
# flush=True → Colab/터미널에서 기다리지 않고 바로 출력.

for token in chain.stream({"topic" : "mutimodel"}):
    print(token, end = "", flush=True)

print("\n")

# 동기 invoke 호출 예시
print("== 동기 invoke 출력 ==")

result = chain.invoke({"topic" : "CahtGPT"})
print(result, "\n")

# 동기 batch 처리 예시
print("== 동기 batch 처리 출력 ==")
# 여러 입력(dict)들을 한 번에 처리하고, 결과를 리스트로 반환.
# 즉, 병렬 처리된 결과 묶음을 얻을 수 있음.

results = chain.batch([{"topic":"ChatGPT"}, {"topic":"Instagram"}])
for r in results:
    print(r)

print()

# 동기 batch with concurrency 예시 : 과부하를 막으면서 병렬처
print("== 동기 batch with concurrency 출력 ==")
# .batch([...]) : 여러 입력을 한 번에 처리 → 리스트 결과 반환
# config={"max_concurrency": N} : 동시에 실행할 개수를 N으로 제한
# 과부하를 막으면서도 병렬 처리가 가능

results = chain.batch(
    [
        {"topic" : "ChatGPT"},
        {"topic" : "Instagram"},
        {"topic" : "멀티모달"},
        {"topic" : "프로그래밍"},
        {"topic" : "머신러닝"}
    ],
    config={"max_concurrency": 3}
)

for r in results:
    print(r)
print()

```

batch + stream 개념
- 여러 요청을 병렬로 실행
- 각 요청의 결과를 생성되는 즉시 출력
- 보통 asyncio + astream() 조합으로 구현

batch
- 여러 입력을 동시에 처리한다.
- 모든 작업이 완료된 후 결과를 리스트로 반환한다.
- 처리 중간 결과는 확인할 수 없다.

stream
- 하나의 요청에 대해 응답을 토큰 단위로 실시간 출력한다.
- 응답 생성 과정을 바로 확인할 수 있다.
- 일반적으로 단일 요청에 사용한다.

```python
# ==========================================
# 기존 코드 (비교용 유지)
# ==========================================

# from langchain_openai import ChatOpenAI
# from langchain.prompts import PromptTemplate
# from langchain.chains import LLMChain
#
# llm = ChatOpenAI(
#     model="gpt-4o-mini",
#     temperature=0.3,
#     streaming=True
# )
#
# prompt = PromptTemplate(
#     input_variables=["topic"],
#     template="Write a short explanation about {topic}."
# )
#
# chain = LLMChain(
#     llm=llm,
#     prompt=prompt
# )

# ==========================================
# 개선 코드 (최신 LCEL 방식)
# ==========================================

from google.colab import userdata

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# OpenAI API Key 로드
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")

# Chat Model 생성
# model : 사용할 OpenAI 모델
# temperature : 생성 다양성 제어
# streaming=True : stream() 사용 시 토큰 단위 출력
llm = ChatOpenAI(
    api_key= OPENAI_API_KEY,
    model = "gpt-4o-mini",
    temperature= 0.3,
    streaming= True
)



# PromptTemplate → ChatPromptTemplate 변경
# Chat Model과 함께 사용할 때 권장되는 방식
prompt = ChatPromptTemplate.from_template(
    "Write a short explanation about {topic}."
)



# LLMChain 제거
# LCEL(Prompt → Model → Parser) 적용
chain = (
    prompt
    | llm
    | StrOutputParser()
)

# 여러 입력 정의
inputs = [
    {"topic" : "ChatGPT"},
    {"topic" : "Instagram"},
    {"topic" : "멀티모달"}
]

print("== Sequential Stream 출력 ==")

# 각 입력에 대해 순차 스트리밍 수행
for inp in inputs:

    print(f"\n--- {inp['topic']} ---")

    # 응답을 토큰 단위로 실시간 출력
    for chunk in chain.stream(inp):
        print(chunk, end="", flush=True)

    print("\n")
```

```python
# ============================================================
# 기존 코드
# ============================================================

# import nest_asyncio
# nest_asyncio.apply()

# ============================================================
# 개선 코드
# ============================================================

# Colab/Jupyter 환경에서는 이미 이벤트 루프가 실행 중이다.
# 따라서 비동기 코드를 실행하기 위해 nest_asyncio를 적용한다.
import nest_asyncio




# ============================================================
# 기존 코드
# ============================================================

# import asyncio

# ============================================================
# 개선 코드
# ============================================================

# 비동기 실행을 위한 asyncio
import asyncio
nest_asyncio.apply()

# ============================================================
# 기존 코드
# ============================================================

# model = ChatOpenAI()

# ============================================================
# 개선 코드
# ============================================================

# OpenAI API Key를 Colab Secrets에서 안전하게 로드
from google.colab import userdata

# 최신 LangChain 권장 Import
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

# Chat Model 생성
# model : 사용할 모델 지정
# temperature : 응답 일관성 확보
model = ChatOpenAI(
    model = "gpt-4o-mini",
    temperature=0
)




# ============================================================
# 기존 코드
# ============================================================

# prompt = PromptTemplate.from_template(
#     '{topic} 에 대해 3문장으로 설명해줘'
# )

# ============================================================
# 개선 코드
# ============================================================

# PromptTemplate → ChatPromptTemplate 변경
# Chat Model과 함께 사용하는 최신 방식
prompt = ChatPromptTemplate.from_template(
    "{topic}에 대해 3문장으로 설명해줘"
)



# ============================================================
# 기존 코드
# ============================================================

# chain = prompt | model

# ============================================================
# 개선 코드
# ============================================================

# LCEL 체인 구성
# Prompt → Model → OutputParser
chain = (
    prompt
    | model
    | StrOutputParser()
)


# ============================================================
# 기존 코드
# ============================================================

# async def async_main():
#
#   print('== 비동기 스트림 출력 ==')
#
#   async for token in chain.astream(
#       {'topic': 'YouTube'}
#   ):
#       print(token, end='', flush=True)
#
#   print('\n')
#
#   print('== 비동기 invoke 출력 ==')
#
#   my_process = chain.ainvoke(
#       {'topic': 'NVDA'}
#   )
#
#   result = await my_process
#
#   print(result,'\n')
#
#   print('== 비동기 batch 출력 ==')
#
#   my_batch = chain.abatch(
#       [
#           {'topic': 'YouTube'},
#           {'topic': 'Instagram'},
#           {'topic': 'Facebook'}
#       ]
#   )
#
#   result = await my_batch
#
#   for r in result:
#       print(r)
#
#   print()

# ============================================================
# 개선 코드
# ============================================================

async def async_main():

    # --------------------------------------------------------
    # astream()
    # --------------------------------------------------------
    print("== 비동기 스트림 출력 ==")

    # 응답을 토큰 단위로 비동기 수신

    async for token in chain.astream(
        {"topic" : "YouTube"}
    ):
        print(token, end="", flush=True)


    print("\n")

    # --------------------------------------------------------
    # ainvoke()
    # --------------------------------------------------------
    print("== 비동기 invoke 출력 ==")

    # 비동기 체인 실행
    result  = await chain.ainvoke (
        {"topic" : "NVIDA"}
    )


    print(result)
    print()

    # --------------------------------------------------------
    # abatch()
    # --------------------------------------------------------
    print("== 비동기 batch 출력 ==")

    # 여러 요청을 병렬 처리
    results = await chain.abatch(
        [
            {"topic":"YouTube"},
            {"topic" : "Instagram"},
            {"topic" : "Facebook"}
        ]
    )



    for r in results:
        print(r)

    print()


# ============================================================
# 기존 코드
# ============================================================

# await async_main()
#
# asyncio.run(async_main())

# ============================================================
# 개선 코드
# ============================================================

# Colab에서는 이미 이벤트 루프가 실행 중이므로
# asyncio.run() 사용 시 오류 발생 가능
#
# RuntimeError:
# asyncio.run() cannot be called from a running event loop
#
# 따라서 await만 사용하는 것이 안전하다.

await async_main()
```

RunnableParallel

- 여러 Runnable을 동시에 실행하기 위한 컴포넌트
- 동일한 입력을 여러 체인에 전달하여 병렬 처리할 수 있다.
- 결과는 딕셔너리(Dictionary) 형태로 반환된다.

특징

1. 여러 체인을 동시에 실행
2. 동일 입력을 여러 체인에 전달 가능
3. 결과를 Key-Value 형태로 반환
4. 실행 시간을 단축할 수 있음

RunnableParallel

- 병렬 요청을 지원하는 Runnable 컴포넌트
- 딕셔너리 형태로 Runnable을 정의
- 각 Runnable은 동일 입력을 받아 동시에 실행
- 결과는 딕셔너리 형태로 반환

Sequential
prompt | model | parser

Parallel
RunnableParallel(
    key1=chain1,
    key2=chain2
)

입력
 ↓

 ┌───────────────┐
 │ Runnable A    │
 └───────────────┘
          │

 ┌───────────────┐
 │ Runnable B    │
 └───────────────┘
          │

 ┌───────────────┐
 │ Runnable C    │
 └───────────────┘

동시에 실행
 ↓

{
  "A": 결과,
  "B": 결과,
  "C": 결과
}

```python

# 여러 Runnable을 동시에 실행하기 위한 컴포넌트
from langchain_core.runnables import RunnableParallel


# ============================================================
# 기존 코드
# ============================================================

# from langchain_core.prompts import PromptTemplate

# ============================================================
# 개선 코드
# ============================================================

# Chat Model과 함께 사용하는 최신 권장 Prompt
from langchain_core.prompts import ChatPromptTemplate


# ============================================================
# 기존 코드
# ============================================================

# chain1 = (
#     PromptTemplate.from_template(
#         '{country}의 수도는 어디야?'
#     )
#     | model
#     | StrOutputParser()
# )

# ============================================================
# 개선 코드
# ============================================================

# 국가의 수도를 질의하는 체인
chain1 = (
    ChatPromptTemplate.from_template(
        "{country}의 수도는 어디야?"
    )
    | model
    | StrOutputParser()
)


# ============================================================
# 기존 코드
# ============================================================

# chain2 = (
#     PromptTemplate.from_template(
#         '{country}의 면적은 얼마야?'
#     )
#     | model
#     | StrOutputParser()
# )

# ============================================================
# 개선 코드
# ============================================================

# 국가의 면적을 질의하는 체인
chain2 = (
    ChatPromptTemplate.from_template(
        "{country}의 면적은 얼마야?"
    )
    | model
    | StrOutputParser()
)


# RunnableParallel
#
# 입력값을 각 체인에 동시에 전달한다.
#
# chain1 → 수도 조회
# chain2 → 면적 조회
#
# 두 체인이 병렬 실행된 후
# 결과를 Dictionary 형태로 반환한다.

combined = RunnableParallel(
    capital = chain1,
    area = chain2
)
```

```python
# ============================================================
# 기존 코드
# ============================================================

# chain1.invoke(
#     {'country':'대한민국'}
# )

# ============================================================
# 개선 코드
# ============================================================

# chain1 단독 실행
#
# 입력값(country)을 받아
# "대한민국의 수도는 어디야?"
# 프롬프트를 생성한 뒤 모델에 전달한다.
#
# 반환값은 StrOutputParser에 의해
# 순수 문자열(String) 형태로 반환된다.

result = chain1.invoke(
    {
        "country": "대한민국"
    }
)

print("수도 정보 :", result)
```

```python
# chain2 실행
result = chain2.invoke(
    {
        "country": "미국"
    }
)

print("면적 정보 :", result)
```

```python

# RunnableParallel 실행
#
# 동일한 입력값(country)을
# chain1(수도 조회)와
# chain2(면적 조회)에 동시에 전달한다.
#
# 실행 결과는 Dictionary 형태로 반환된다.

result = combined.invoke(
    {"country" : "대한민국"}
)



# 전체 결과 출력
print("병렬 실행 결과 :")
print(result)

print()

# 결과 개별 출력
print("수도 정보 :")
print(result["capital"])

print()

print("면적 정보 :")
print(result["area"])
```

batch()에서의 병렬 처리

- batch()는 여러 개의 입력 데이터를 한 번에 처리하기 위한 메서드이다.
- 여러 개의 딕셔너리를 포함하는 리스트(List)를 입력으로 받는다.
- 각 딕셔너리는 체인에 전달될 입력값을 나타낸다.
- 각 입력은 독립적으로 처리되며 결과는 리스트(List) 형태로 반환된다.
- 내부적으로 병렬 실행될 수 있어 여러 요청을 효율적으로 처리할 수 있다.

```python

# Runnable의 batch()
#
# 여러 입력을 한 번에 처리한다.
# 각 입력은 독립적으로 chain1에 전달된다.
#
# 반환값은 결과들의 List이다.

results = chain1.batch(
    [
        {"country" : "대한민국"},
        {"country" : "미국"}
    ]
)



# 전체 결과 출력
print("batch 결과 :")
print(results)

print()

# 결과 개별 출력
for country, result in zip(
    ["대한민국", "미국"],
    results
):
    print(f"{country} 수도 정보 : ")
    print(result)
    print()
```

```python

# Runnable의 batch()
#
# 여러 국가 정보를 한 번에 처리한다.
# chain2는 국가의 면적을 질의하는 체인이다.
#
# 반환값은 결과들의 List 형태이다.

results = chain2.batch(
    [
        {"country" : "대한민국"},
        {"country" : "미국"}
    ]
)




# 전체 결과 확인
print("batch 결과 :")
print(results)

print()

# 결과 개별 출력
for country, result in zip(
    ["대한민국", "미국"],
    results
):
    print(f"{country} 면적 정보 :")
    print(result)
    print()
```

```python


# RunnableParallel + batch()
#
# batch()는 여러 입력을 처리한다.
#
# combined는 RunnableParallel이므로
# 각 입력마다
#
# 1. 수도 조회(chain1)
# 2. 면적 조회(chain2)
#
# 를 동시에 수행한다.
#
# 반환 결과는 List[Dict] 형태이다.

results = combined.batch(
    [
        {"country" : "대한민국"},
        {"country" : "미국"}
    ]
)



# 전체 결과 확인
print("배치 처리 결과 :")
print(results)

print()

# 결과 상세 출력
for country, result in zip(
    ["대한민국", "미국"],
    results
):
    print(f"===== {country} =====")

    print("수도 정보 :")
    print(result["capital"])

    print()

    print("면적 정보 :")
    print(result["area"])

    print()
```

데이터를 효과적으로 전달하는 방법

RunnablePassthrough

- 입력 데이터를 변경하지 않고 그대로 다음 Runnable에 전달할 수 있다.
- 기존 입력값을 유지하면서 새로운 데이터를 추가할 수도 있다.
- LCEL에서 입력값을 재사용하거나 중간 결과를 함께 전달할 때 자주 사용된다.

주요 기능

1. RunnablePassthrough()

   - 입력값을 그대로 반환한다.
   - 입력 데이터를 수정하지 않고 다음 단계로 전달할 때 사용한다.

2. RunnablePassthrough.assign()

   - 기존 입력 데이터를 유지한다.
   - 새로운 Key-Value 데이터를 추가한다.
   - 여러 체인의 결과를 하나의 Dictionary로 합칠 때 유용하다.

RunnableLambda로 동적 분기

```python
# ------------------------------------------------------------
# [상세 설명]
# 이 코드는 사용자의 질문을 "수학 / 과학 / 기타"로 분류한 뒤
# 해당 카테고리에 맞는 전문가 체인으로 라우팅하는 LCEL 구조이다.
#
# 전체 흐름:
# 1. classification_chain → 질문 분류
# 2. router() → 분류 결과 기반 체인 선택
# 3. math / science / general chain → 응답 생성
# 4. RunnableLambda → 동적 실행 구조 구성
# ------------------------------------------------------------

import os
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda
from langchain_openai import ChatOpenAI

# OpenAI API 키 설정
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get("OPENAI_API_KEY")


# 1. 질문 분류 체인 (수학 / 과학 / 기타)
classification_prompt = PromptTemplate.from_template(
    """주어진 사용자 질문을 `수학`, `과학`, 또는 `기타` 중 하나로 분류하세요. 한 단어 이상으로 응답하지 마세요.

<question>
{question}
</question>

Classification:"""
)

classification_chain = (
    classification_prompt
    | ChatOpenAI(model="gpt-4o", temperature=0)
    | StrOutputParser()
)


# 2. 수학 체인
math_chain = (
    PromptTemplate.from_template(
        """You are an expert in math. \
Always answer questions starting with "깨봉선생님께서 말씀하시기를..". \
Respond to the following question:

Question: {question}
Answer:"""
    )
    | ChatOpenAI(model="gpt-4o", temperature=0)
    | StrOutputParser()
)


# 3. 과학 체인
science_chain = (
    PromptTemplate.from_template(
        """You are an expert in science. \
Always answer questions starting with "아이작 뉴턴 선생님께서 말씀하시기를..". \
Respond to the following question:

Question: {question}
Answer:"""
    )
    | ChatOpenAI(model="gpt-4o", temperature=0)
    | StrOutputParser()
)


# 4. 기타 체인
general_chain = (
    PromptTemplate.from_template(
        """Respond to the following question concisely:

Question: {question}
Answer:"""
    )
    | ChatOpenAI(model="gpt-4o", temperature=0)
    | StrOutputParser()
)


# 5. 라우터 함수: 분류 결과에 따라 적절한 체인 선택
def router(input: dict):
    category = classification_chain.invoke(
        {"question" : input["question"]}
    ).strip()

    if category == "수학":
        return math_chain
    elif category == "과학":
        return science_chain
    else:
        return general_chain



# 6. 최종 라우팅 체인 (RunnableLambda로 동적 분기)
router_chain = RunnableLambda(
    lambda input : router(input).invoke(input)
)



# 7. 테스트 실행
questions = [
    "2+2는 무엇인가요?",
    "작용 반작용 법칙은 무엇인가요?",
    "Google은 어떤 회사인가요?",
]

for q in questions:
    print(f"\n질문: {q}")
    answer = router_chain.invoke({"question": q})
    print(f"답변: {answer}")
```

런타임에 체인 내부 구성하기

- 이번 튜토리얼은 Chain을 호출할 때 다양한 옵션을 동적으로 설정하는 방법을 설명한다.

- LangChain에서는 실행 시점에 체인의 동작을 변경할 수 있으며, 이를 통해 코드 수정 없이 다양한 실험과 구성 변경이 가능하다.

1. configurable_fields

  - 실행 가능한 객체의 특정 필드를 런타임에서 변경할 수 있는 기능이다.

  - 예를 들어 다음과 같은 값을 동적으로 변경할 수 있다:

    1. 모델 종류 (GPT-4o, GPT-3.5 등)
    2. temperature 값
    3. 프롬프트 내부 변수

  - 즉, 체인의 구조는 그대로 두고 내부 설정만 바꾸는 방식이다.

2. configurable_alternatives
  - 실행 중에 구성 요소 자체를 교체할 수 있는 기능이다.

  - 예를 들어:

    1. 서로 다른 LLM 교체 (GPT ↔ Claude 등)

    3. 체인 구성 요소 전체 변경

  - 즉, 단순 파라미터 변경이 아니라 실행 구조 자체를 바꾸는 방식이다.


상세 설명


- @chain 데코레이터는 LangChain에서 함수를 Runnable 객체로 변환하는 기능이다.

- 즉, 일반 Python 함수를 LCEL 체인에 바로 연결할 수 있게 만들어준다.

- 핵심 개념
1. 일반 함수 → RunnableLambda 자동 변환
2. LCEL 체인에 직접 연결 가능
3. 복잡한 클래스 없이 간단한 파이프라인 구성 가능
- 핵심 장점
1. 코드가 매우 간결해진다
2. 함수 기반 로직을 그대로 체인에 연결 가능
3. RunnableLambda보다 더 직관적
4. 디버깅이 쉬움
5. LCEL 구조와 자연스럽게 통합됨

```python
#============================
""" 기존 코드 vs 개선 코드
기존 방식:
- RunnableLambda(add_one)
- 클래스 기반 구조
- 코드 길어짐

개선 방식 (@chain):
- 함수 위에 데코레이터
- 자동 Runnable 변환
- LCEL과 자연스럽게 연결
"""
#============================


#2-1. RunnableLambda 방식 (기존 방식)
from langchain_core.runnables import RunnableLambda

def add_one(x):
    return x + 1


chain = RunnableLambda(add_one)

result = chain.invoke(10)
print(result)
```

```python
#2-2. @chain 데코레이터 방식 (개선)
from langchain_core.runnables import chain

# 함수 자체를 Runnable로 변환

@chain
def add_one(x):
    return x + 1

result = add_one.invoke(10)

print(result)
```

```python
#3. LCEL과 함께 사용하는 방식

from langchain_core.runnables import chain

@chain
def preprocess(x):
    return x*2

@chain
def postprocess(x):
    return f"결과값 : {x}"

# 체인 연결 (함수 기반 LCEL)
final_chain = preprocess | postprocess

result = final_chain.invoke(5)
print(result)
```

ChatPromptTemplate 클래스를 사용하여 두 개의 프롬프트 템플릿을 정의합니다.

- prompt1 은 주어진 주제에 대한 짧은 설명을, - - prompt2 는 영어로 번역해 달라는 요청 프롬프트 입니다.

custom_chain 함수는 입력 텍스트를 기반으로 사용자 정의 체인을 실행합니다.

- @chain 데코레이터로 사용자 정의 함수를 데코레이팅 하며, 데코레이팅을 통해 함수를 Runnable 한 객체로 만듭니다.

```python
# ============================================================
# 1. 상세 설명
# ============================================================

# 이 코드는 @chain 데코레이터를 사용하여
# 두 개의 LCEL 체인을 순차적으로 연결하는 구조다.

# 전체 흐름:
# 입력 text
#   → prompt1 + LLM (요약)
#   → prompt2 + LLM (SNS 변환)
#   → 최종 결과 반환

# 핵심:
# - 체인 내부에서 또 다른 체인을 호출하는 구조
# - 순차적 파이프라인 (sequential pipeline)
# - LCEL + @chain 결합 패턴

# ============================================================
# 2. 필수 import
# ============================================================

from langchain_core.runnables import chain  #chain 데코레이터
from langchain_openai import ChatOpenAI     # LLM
from langchain_core.output_parsers import StrOutputParser  # 문자열 변환

# ============================================================
# 3. 프롬프트 (기존 코드 유지)
# ============================================================

prompt1 = ChatPromptTemplate.from_template(
    "{topic} 에 대해 짧게 한글로 설명해주세요."
)

prompt2 = ChatPromptTemplate.from_template(
    "{sentence} 를 emoji를 활용한 인스타그램 게시글로 만들어주세요."
)

# ============================================================
# 4. 기존 코드 vs 개선 코드 (@chain 적용)
# ============================================================

# ----------------------------
# 기존 코드
# ----------------------------

# def custom_chain(text):
#     chain1 = prompt1 | ChatOpenAI(model="gpt-4o-mini") | StrOutputParser()
#     output1 = chain1.invoke({"topic": text})
#
#     chain2 = prompt2 | ChatOpenAI(model="gpt-4o-mini") | StrOutputParser()
#     return chain2.invoke({"sentence": output1})


# ----------------------------
# 개선 코드 (@chain 적용)
# ----------------------------

@chain
def custom_chain(text):

    # ============================================================
    # 1단계: 설명 생성 체인
    # ============================================================

    chain1 = (
        prompt1
        | ChatOpenAI(model = "gpt-4o-mini")
        | StrOutputParser()
    )


    # 첫 번째 체인 실행
    output1 = chain1.invoke({"topic" : text})


    # ============================================================
    # 2단계: SNS 변환 체인
    # ============================================================

    chain2 = (
        prompt2
        | ChatOpenAI(model="gpt-4o-mini")
        | StrOutputParser()
    )


    # 두 번째 체인 실행 (1단계 결과 입력)
    final_output = chain2.invoke({"sentence" : output1})


    # 최종 결과 반환
    return final_output


# ============================================================
# 5. 실행 테스트
# ============================================================

result = custom_chain.invoke("인공지능")


print("최종 결과:\n", result)
```

custom_chain은 이제 실행 가능한 객체(runnable)이므로, invoke() 를 사용하여 실행해야 합니다.

**`langchain`** 라이브러리를 활용한 파이프라인 처리 시스템입니다. 이 시스템은 **텍스트 전처리**, **요약**, **GPT 모델을 활용한 질의응답**의 세 가지 단계를 거쳐, 최종적으로 **GPT-4** 모델을 사용하여 답변을 생성하는 구조

### **핵심 개념**

1. **Runnable**: `langchain`에서 각 단계를 `Runnable`로 정의하여, 그 각 단계가 독립적으로 작동하고, 그 결과를 다른 단계로 전달할 수 있도록 합니다.
2. **@chain**: `@chain` 데코레이터는 여러 `Runnable` 객체를 연결하여, 순차적으로 처리할 수 있게 해주는 역할을 합니다. 즉, 여러 단계를 체인처럼 이어서 실행할 수 있게 합니다.

### 왜 이 코드는 잘 실행되는지:

* `@chain`은 각 단계가 순차적으로 연결되고, `invoke` 메서드가 호출되면 각 객체가 연결된 순서대로 실행됩니다.
* `CleanText`, `SummarizeText`, `AskGPT`의 각 객체들은 그 자체로 `Runnable` 객체이므로, `invoke` 메서드를 호출하여 그 흐름을 이어가는 것이 가능합니다.
* 각 단계는 `invoke` 메서드를 통해 데이터를 주고받으며, 마지막에 `AskGPT` 객체에서 최종 결과가 생성됩니다.

```python
"""
============================================================
LangChain 파이프라인 처리 시스템 설명
============================================================

이 시스템은 LangChain을 활용하여 텍스트 전처리 → 요약 → GPT 기반 질의응답으로 이어지는
3단계 파이프라인 구조를 구성한 것이다.

전체 구조는 각 단계가 독립적인 실행 단위로 동작하면서도, 결과를 다음 단계로 전달하는 방식으로
설계되어 있다.

============================================================
핵심 개념
============================================================
1. Runnable

LangChain에서 각 처리 단계는 Runnable로 정의된다.

Runnable의 특징은 다음과 같다:

각 단계가 독립적으로 실행 가능
invoke()를 통해 입력을 처리
출력 결과를 다음 Runnable로 전달 가능
체인 구조에서 재사용 가능

즉, 모든 단계는 “입력 → 처리 → 출력” 구조를 갖는 독립 실행 단위이다.

2. @chain 데코레이터

@chain은 여러 Runnable 객체를 하나의 흐름으로 연결하는 역할을 한다.

핵심 역할은 다음과 같다:

함수 또는 로직을 Runnable 체인으로 변환
여러 단계를 순차적으로 연결
invoke() 호출 시 전체 파이프라인 실행

즉, 여러 단계의 Runnable을 “파이프라인 구조”로 실행할 수 있게 한다.

============================================================
왜 이 코드가 정상 동작하는가
============================================================

이 구조가 정상적으로 동작하는 이유는 다음과 같다:

1. 모든 단계가 Runnable 구조

CleanText, SummarizeText, AskGPT 각각은 독립적인 Runnable 객체로 구성되어 있다.

따라서 각 단계는 invoke() 메서드를 통해 실행 가능하다.

2. @chain에 의해 자동 연결

@chain 데코레이터는 내부적으로 각 Runnable을 순차적으로 연결한다.

즉:

CleanText → SummarizeText → AskGPT

형태의 실행 흐름이 자동으로 구성된다.

3. invoke() 기반 데이터 흐름

각 단계는 invoke()를 통해 데이터를 전달받고 처리 결과를 반환한다.

입력 데이터는 첫 번째 단계로 전달됨
각 단계의 출력은 다음 단계의 입력이 됨
마지막 단계의 출력이 최종 결과가 됨
4. 최종 출력 생성

마지막 단계인 AskGPT는 GPT-4 모델을 활용하여 최종 응답을 생성한다.

따라서 전체 파이프라인의 결과는 GPT-4 기반 응답이 된다.

============================================================
전체 구조 요약
============================================================
입력 텍스트
   ↓
CleanText (전처리)
   ↓
SummarizeText (요약)
   ↓
AskGPT (GPT-4 질의응답)
   ↓
최종 결과 출력
============================================================
핵심 정리
============================================================
Runnable: 각 단계의 독립 실행 단위
@chain: Runnable을 순차적으로 연결하는 구조
invoke(): 단계 간 데이터 전달 메커니즘
전체 구조: 파이프라인 기반 LLM 처리 시스템

요약
LangChain의 @chain 기반 파이프라인은 각 Runnable을 독립 실행 단위로 유지하면서,
invoke 기반 데이터 흐름으로 순차 처리되는 구조이다.
"""
```

```python
# ============================================================
# 1. 상세 설명
# ============================================================

# 이 코드는 LangChain의 Runnable과 @chain을 활용하여
# 텍스트 처리 파이프라인을 구성한 예제이다.

# 전체 흐름:
# 입력 텍스트
#   → CleanText (전처리)
#   → SummarizeText (요약)
#   → AskGPT (GPT-4 응답 생성)
#   → 최종 결과 반환

# 핵심 개념:
# - Runnable: 각 단계 독립 실행 단위
# - @chain: Runnable을 순차적으로 연결
# - invoke(): 데이터 전달 및 실행 트리거

# ============================================================
# 2. 필수 import
# ============================================================

from langchain_core.runnables import Runnable, chain  # Runnable + chain 데코레이터
from langchain_core.prompts import PromptTemplate     # 프롬프트 템플릿
from langchain_openai import ChatOpenAI               # GPT 모델

# ============================================================
# 3. 프롬프트 정의
# ============================================================

# GPT에게 전달할 질문 형태 정의
prompt = PromptTemplate.from_template(
    "이 텍스트를 기반으로 답변을 생성하세요: {data}"
)

# ============================================================
# 4. Runnable 단계 정의
# ============================================================

# ----------------------------
# 1단계: 텍스트 전처리
# ----------------------------
class CleanText(Runnable):
    def invoke(self, data: str) -> str:
        return data.strip().lower()


# ----------------------------
# 2단계: 텍스트 요약
# ----------------------------
class SummarizeText(Runnable):
    def invoke(self, data: str) -> str:
        return data[:100] + "..."



# ----------------------------
# 3단계: GPT 질의응답
# ----------------------------
class AskGPT(Runnable):
    def invoke(self, data: str) -> str:
        model = ChatOpenAI(
            temperature = 0,
            model = "gpt-4"
        )

        response = model.invoke(prompt.format(data=data))
        return response.content



# ============================================================
# 5. @chain 기반 파이프라인 구성
# ============================================================

@chain
def custom_chain(data: str) -> str:

    # 1단계: 전처리
    clean_txt = CleanText()
    processed_txt = clean_txt.invoke(data)


    # 2단계: 요약
    summarize_txt = SummarizeText()
    summarized_txt = clean_txt.invoke(processed_txt)


    # 3단계: GPT 응답 생성
    ask_gpt = AskGPT()
    final_result = ask_gpt.invoke(summarized_txt)

    return final_result



# ============================================================
# 6. 테스트 실행
# ============================================================

sample_data = """
OpenAI의 GPT-4 모델은 자연어 처리(NLP)에서 가장 최신 기술 중 하나입니다.
이 모델은 다양한 언어 이해 및 생성 능력을 가지고 있으며,
질문 응답, 요약, 번역 등 다양한 작업을 수행할 수 있습니다.
"""

result = custom_chain.invoke(sample_data)

print("결과:", result)
```

```python
# ============================================================
# 1. 상세 설명
# ============================================================

# 이 코드는 LangChain Runnable을 활용하여
# 텍스트 전처리 → 요약 → GPT 질의응답을 수행하는 파이프라인 구조이다.

# 각 단계는 Runnable로 구현되어 있으며,
# invoke()를 통해 순차적으로 데이터가 전달된다.

# 전체 흐름:
# 입력 텍스트
#   → CleanText (전처리)
#   → SummarizeText (요약)
#   → AskGPT (GPT-4 응답 생성)
#   → 최종 결과 반환

# ============================================================
# 2. 필수 import
# ============================================================

from langchain_core.runnables import Runnable  # Runnable 인터페이스
from langchain_core.prompts import PromptTemplate  # 프롬프트 템플릿
from langchain_openai import ChatOpenAI  # GPT 모델

# ============================================================
# 3. 프롬프트 정의
# ============================================================

# 입력 데이터를 기반으로 GPT에게 전달할 프롬프트 생성
prompt = PromptTemplate.from_template(
    "이 텍스트를 기반으로 답변을 생성하세요: {data}"
)

# ----------------------------
# 1단계: 텍스트 전처리
# ----------------------------
class CleanText(Runnable):
    def invoke(self, data: str) -> str:
        return data.strip().lower()


# ----------------------------
# 2단계: 텍스트 요약
# ----------------------------
class SummarizeText(Runnable):
    def invoke(self, data: str) -> str:
        return data[:100] + "..."



# ----------------------------
# 3단계: GPT 질의응답
# ----------------------------
class AskGPT(Runnable):
    def invoke(self, data: str) -> str:
        model = ChatOpenAI(
            temperature = 0,
            model = "gpt-4"
        )

        response = model.invoke(prompt.format(data=data))
        return response.content




# ============================================================
# 7. 4단계: 파이프라인 구성 클래스
# ============================================================

class ProcessPipeline(Runnable):

    def __init__(self):

        # 전처리 Runnable 생성
        self.clean_txt = CleanText()

        # 요약 Runnable 생성
        self.summarize_txt = SummarizeText()


        # GPT 호출 Runnable 생성
        self.ask_gpt = AskGPT()


    def invoke(self, data: str) -> str:

        # ========================================================
        # 1단계: 텍스트 전처리 실행
        # ========================================================
        data = self.clean_txt.invoke(data)


        # ========================================================
        # 2단계: 텍스트 요약 실행
        # ========================================================
        data = self.summarize_txt.invoke(data)


        # ========================================================
        # 3단계: GPT 응답 생성 실행
        # ========================================================
        data = self.ask_gpt.invoke(data)


        # 최종 결과 반환
        return data

# ============================================================
# 8. 테스트 실행
# ============================================================

# 입력 데이터 정의
sample_data = """
OpenAI의 GPT-4 모델은 자연어 처리(NLP)에서 가장 최신 기술 중 하나입니다.
이 모델은 다양한 언어 이해 및 생성 능력을 가지고 있으며,
질문 응답, 요약, 번역 등 다양한 작업을 수행할 수 있습니다.
"""

# 파이프라인 객체 생성
pipeline = ProcessPipeline()


# 전체 파이프라인 실행
result = pipeline.invoke(sample_data)


# 결과 출력
print("결과:", result)
```
