```python
import pandas as pd
import numpy as np
import os
```

```python
path_dir = 'D:/ml/statistics-for-data-scientists/c_data/'
```

```python
file_list = os.listdir(path_dir)
```

```python
file_list
```

```python
for file_name in file_list:
    
    
    print(file_name)
    
    
    df = pd.read_excel(path_dir + str(file_name))
    
    set_username = set(list(df['username']))    
    
    for i in set_username:
        
        #print(i)
    
        tweet_count = int(len(df[df['username'] == i]))
        text_count = int(len(df[df['username'] == i]['text_t'].value_counts()))
        tweet_text_ratio = round(text_count / tweet_count, 2) 



        #print (text_count)

        #if tweet_count > text_count & int(tweet_count) > 100:
        if tweet_count > 10 and tweet_count > text_count and tweet_text_ratio < 0.3:
        #if text_count > 100:

            print("{0}  {1}  {2}  {3}".format(i, tweet_count, text_count, tweet_text_ratio))

            ad_user_idx_nm = df[df['username']==i].index

            df.drop(ad_user_idx_nm, inplace = True)


        else :

            pass
        
        
    df.to_excel(path_dir+"ad_user_"+str(file_name))
```
