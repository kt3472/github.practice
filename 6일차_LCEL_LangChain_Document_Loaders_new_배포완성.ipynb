문서 로더
  - LangChain의 문서로더를 사용해서 다양한 형식의 데이터 파일로 부터 문서로 로드할 수 있다
  - Document 객체로 로드되고, page_content에는 내용이 들어가고, metadata에는 메타데이터를 포함한다
  - txt 형식의 문서 뿐만아니라 pdf, word, ppt, xlsx,csv, hwp등 거의 문든 형식으 문서를 기반으로 LLM을 구동할 수 있다
  - 기능
    1. load()
      - 문서를 로드하여 반환한다
      - 반환된 결과는 List[] 형태이다
    2. load_and_split()
      - splitter를 사용하여 문서를 분할하고 반환한다
      - 반환된 결과는 List[]
    3. lazy_load()
      - generator 방식으로 문서를 로드한다
    4. aload()
      - 비동기식(Async)방식의 문서로

```python
# PyMuPDF(fitz)를 활용한 텍스트 추출 방식은 PyPDFLoader보다
# 훨씬 빠르고 정확하다는 장점이 있어 실무에서 아주 많이 사용

import fitz # PyMuPDF 라이브러리 임포트 (fitz는 PyMuPDF의 별칭)
import requests
import os

# PDF 파일에서 텍스트를 추출하는 함수 정의
def extract_text_from_pdf(pdf_path):

  # PDF 파일 열기
  document = fitz.open(pdf_path)

  # 추출된 텍스트를 저장할 변수 초기화
  text = ''

  # PDF의 각 페이지를 반복 처리
  for page_num in range(len(document)):
    page = document.load_page(page_num)

    text += page.get_text()

  return text

# 예제 PDF 파일 URL 및 로컬 저장 경로 설정
pdf_url = 'https://pdfobject.com/pdf/sample.pdf'

local_pdf_path = 'sample.pdf'

# URL에서 PDF 파일 다운로드 요청
response = requests.get(pdf_url)


# 다운로드 받은 콘텐츠를 로컬 파일로 저장 (바이너리 쓰기 모드)
with open(local_pdf_path, 'wb') as f:
    f.write(response.content)


# PDF 파일에서 텍스트 추출 함수 호출
extract_text = extract_text_from_pdf(local_pdf_path)


# 추출된 텍스트 출력
print(extract_text)

# 다운로드한 로컬 PDF 파일 삭제
os.remove(local_pdf_path)
```

```python
# python-pptx 라이브러리 설치 (PPTX 텍스트 추출을 위해 사용)
!pip install -q python-pptx
```

```python
from pptx import Presentation
from google.colab import files # Colab 파일 업로드를 위해 임포트

# PPTX 파일에서 텍스트를 추출하는 함수 정의
def extract_text_from_pptx(pptx_path):

  # Presentation 객체 생성 (PPTX 파일 로드)
  presentation = Presentation(pptx_path)

  # 추출된 텍스트를 저장할 변수 초기화
  text = ''

  # 각 슬라이드를 반복 처리
  for slide in presentation.slides:
    # 슬라이드 내의 각 도형을 반복처리
    for shape in slide.shapes:
      # 도형안에 텍스트 프레임이 있는지 확인
      if shape.has_text_frame:
        # 텍스트 프레임 내의 각 단락을 반복 처리
        for paragraph in shape.text_frame.paragraphs:
          # 각 단락의 텍스트를 추출하여 추가
          text += paragraph.text + '\n'

  return text

# Colab 파일 업로드 기능 실행 (사용자가 파일을 선택하여 업로드)
# 업로드된 파일은 현재 디렉토리에 저장됩니다.
uploaded = files.upload()

# 예제 pptx 파일 경로 설정 (업로드된 파일 이름과 일치해야 함)
# 업로드된 파일 목록에서 첫 번째 파일 이름을 가져와 사용
if uploaded:
    pptx_path = list(uploaded.keys())[0]
    # PPTX 파일에서 텍스트 추출 함수 호출

    extract_text = extract_text_from_pptx(pptx_path)
    # 추출된 텍스트 출력
    print(extract_text)
else:
    print("PPTX 파일이 업로드되지 않았습니다.")
```

```python
# python-docx 라이브러리 설치 (DOCX 텍스트 추출을 위해 사용)
!pip install -q python-docx
```

```python
from docx import Document
from google.colab import files # Colab 파일 업로드를 위해 임포트

# DOCX 파일에서 텍스트를 추출하는 함수 정의
def extract_text_from_docx(docx_path):
  # Document 객체 생성 (DOCX 파일 로드)
  document = Document(docx_path)

  # 추출된 텍스트를 저장할 변수 초기화
  text = ''

  # 문서의 각 단락을 반복 처리
  for para in document.paragraphs:
    text += para.text + '\n' # 각 단락의 텍스트와 개행 문자를 추가
  return text

# Colab 파일 업로드 기능 실행 (사용자가 파일을 선택하여 업로드)
# 업로드된 파일은 현재 디렉토리에 저장됩니다.
uploaded = files.upload()

# 예제 docx 파일 경로 설정 (업로드된 파일 이름과 일치해야 함)
# 업로드된 파일 목록에서 첫 번째 파일 이름을 가져와 사용
if uploaded:
    docx_path = list(uploaded.keys())[0]

    # DOCX 파일에서 텍스트 추출 함수 호출
    extract_text = extract_text_from_docx(docx_path)

    # 추출된 텍스트 출력
    print(extract_text)
else:
    print("DOCX 파일이 업로드되지 않았습니다.")
```

```python
# openpyxl 라이브러리 설치 (XLSX 텍스트 추출을 위해 사용)
!pip install -q openpyxl
```

```python
import openpyxl
from google.colab import files # Colab 파일 업로드를 위해 임포트

# XLSX 파일에서 텍스트를 추출하는 함수 정의
def extract_text_from_xlsx(xlsx_path):

  # Workbook 객체 생성 (XLSX 파일 로드)
  workbook = openpyxl.load_workbook(xlsx_path)
  # 활성화된 시트 선택
  sheet = workbook.active

  # 추출된 텍스트를 저장할 변수 초기화
  text = ''

  # 시트의 각 행을 반복 처리
  for row in sheet.iter_rows(values_only=True):  # 셀값만 가져옴
    # 각 행의 셀값을 문자열로 변환 후 쉼표로 연결하여 텍스트에 추가
    text += ", ".join([str(cell) for cell in row]) + '\n'

  return text

# Colab 파일 업로드 기능 실행 (사용자가 파일을 선택하여 업로드)
# 업로드된 파일은 현재 디렉토리에 저장됩니다.
uploaded = files.upload()

# 예제 xlsx 파일 경로 설정 (업로드된 파일 이름과 일치해야 함)
# 업로드된 파일 목록에서 첫 번째 파일 이름을 가져와 사용
if uploaded:
    xlsx_path = list(uploaded.keys())[0]

    # XLSX 파일에서 텍스트 추출 함수 호출
    extract_text = extract_text_from_xlsx(xlsx_path)
    # 추출된 텍스트 출력
    print(extract_text)
else:
    print("XLSX 파일이 업로드되지 않았습니다.")
```

```python
import olefile   # OLE 구조 파일 처리 (HWP 파일 처리에 사용)
import zlib      # 압축 해제 (HWP 파일 압축 해제에 사용)
import struct    # 바이너리 데이터 처리 (HWP 파일 구조체 분석에 사용)
import re        # 정규 표현식 (추출된 텍스트 정제에 사용)
from google.colab import files # Colab 파일 업로드를 위해 임포트

# HWP 파일에서 텍스트를 추출하는 함수 정의
def get_hwp_text(filename):
  # OLE 파일 열기
  f = olefile.OleFileIO(filename)
  dirs = f.listdir()  # 파일 내부 디렉토리 구조 가져오기

  # 유효한 HWP 파일인지 확인 (필수 스트림 존재 여부 확인)
  if ['FileHeader'] not in dirs or ['\x05HwpSummaryInformation'] not in dirs:
    raise Exception('Not Valid HWP')

  # FileHeader 스트림 열기 및 데이터 읽기
  header = f.openstream('FileHeader')
  header_data = header.read()
  # 압축 여부 확인 (헤더의 특정 바이트 확인)
  is_compressed = (header_data[36] & 1)

  # 본문 섹션 번호 저장할 리스트 초기화
  nums = []
  # BodyText 섹션을 찾아 번호 추출
  for d in dirs:
    if d[0] == 'BodyText':
      nums.append(int(d[1][len('Section'):]))

  # 섹션 번호를 오름차순으로 정렬하고 섹션 경로 리스트 생성
  sections = ['BodyText/Section' + str(x) for x in sorted(nums)]

  # 추출된 전체 텍스트를 저장할 변수 초기화
  text = ''
  # 각 본문 섹션 반복 처리
  for section in sections:
    # BodyText 섹션 스트림 열기
    bodytext = f.openstream(section)
    data = bodytext.read()
    # 압축 해제 (압축된 경우)
    if is_compressed:
      unpacked_data = zlib.decompress(data, -15) # zlib 압축 해제
    else:
      unpacked_data = data

    # 현재 섹션의 텍스트를 저장할 변수 초기화
    section_text = ''
    i = 0
    size = len(unpacked_data)

    # 섹션 데이터 파싱하여 텍스트 추출
    while i < size:
      # 레코드 헤더 정보 읽기 (구조체 언팩)
      # 4바이트가 남아있는지 확인
      if i + 4 > size:
          break
      # little-endian long (4 bytes)으로 언팩
      header = struct.unpack_from('<l', unpacked_data, i)[0]
      rec_type = header & 0x3ff   # 레코드 타입 추출 (하위 10비트)
      rec_len = (header >> 20) & 0xfff   # 레코드 길이 추출 (20비트부터 12비트)

      # 레코드 타입이 텍스트 관련(67)인 경우 처리 (HWP 문서 형식에 따라 다를 수 있음)
      if rec_type in [67]:
        # 레코드 데이터를 읽을 만큼의 바이트가 남아있는지 확인
        if i + 4 + rec_len > size:
            break
        # 레코드 데이터 추출
        rec_data = unpacked_data[i + 4 : i + 4 + rec_len]
        try:
            # utf-16으로 디코딩하여 섹션 텍스트에 추가
            section_text += rec_data.decode('utf-16')
            #section_text += '\n' # 필요에 따라 개행 추가 (HWP 구조에 따라 다름)
        except UnicodeDecodeError:
            # 디코딩 오류 발생 시 건너뛰기
            pass

      # 다음 레코드로 이동 (헤더 길이 4 + 레코드 길이)
      i += 4 + rec_len

    # 현재 섹션의 텍스트를 전체 텍스트에 추가
    text += section_text
    text += '\n' # 섹션 구분자로 개행 추가
  return text

# Colab 파일 업로드 기능 실행 (사용자가 파일을 선택하여 업로드)
# 업로드된 파일은 현재 디렉토리에 저장됩니다.
uploaded = files.upload()

# 업로드된 파일 목록에서 첫 번째 파일 이름을 가져와 HWP 파일 경로로 설정
if uploaded:
    hwp_path = list(uploaded.keys())[0]
    # HWP 파일에서 텍스트 추출 함수 호출
    text = get_hwp_text(hwp_path)

    # 추출된 텍스트에서 ASCII 및 한글(AC00-D7AF)이 아닌 문자 제거하여 정제
    # 제어 문자 및 기타 불필요한 문자 제거
    cleaned_text = re.sub(r'[^\x20-\x7E\uAC00-\uD7AF\n\r]', '', text) # ASCII 출력 가능 문자, 한글, 개행, 캐리지 리턴만 남김
    # cleaned_text = re.sub(r'[^\x00-\x7F\uAC00-\uD7AF\n\r]', '', text) # ASCII 전체, 한글, 개행, 캐리지 리턴만 남김 (필요에 따라 사용)

    # 정제된 텍스트 출력
    print(cleaned_text)
else:
    print("HWP 파일이 업로드되지 않았습니다.")
```

문제) PDF + Word + CSV 자동 판별기

```python

# ============================================================
#
# [학습 목표]
# 1. 다양한 Document Loader를 사용할 수 있다.
# 2. 파일 확장자에 따라 적절한 Loader를 선택할 수 있다.
# 3. LangChain LCEL을 이용하여 문서를 요약할 수 있다.
#
# ============================================================
# 문제
# ============================================================
#
# 업로드한 파일의 확장자를 확인하여
#
# ① PDF → PyPDFLoader
# ② Word(docx) → Docx2txtLoader
# ③ CSV → CSVLoader
#
# 를 자동으로 선택하여 문서를 읽고
# AI를 이용하여 핵심 내용을 5줄 이내로 요약하는 프로그램을 작성하세요.
#
# ============================================================
# 요구사항
# ============================================================
#
# 1. 파일 확장자를 확인하세요.
# 2. 확장자에 맞는 Document Loader를 생성하세요.
# 3. 문서를 읽으세요.
# 4. PromptTemplate을 작성하세요.
# 5. ChatOpenAI 객체를 생성하세요.
# 6. LCEL Chain을 작성하세요.
# 7. 문서를 요약하여 출력하세요.
#
# ============================================================
# 입력 예시
# ============================================================
#
# report.pdf
#
# 또는
#
# meeting.docx
#
# 또는
#
# student.csv
#
# ============================================================
# 출력 예시
# ============================================================
#
# ==============================
# 문서 요약
# ==============================
#
# - ...
# - ...
# - ...
# - ...
# - ...
#
# ============================================================

import os

from langchain_community.document_loaders import (
    PyPDFLoader,
    Docx2txtLoader,
    CSVLoader,
)

from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

# ============================================================
# TODO 1
#
# 분석할 파일 경로를 저장하세요.
#
# 예)
# file_path = "report.pdf"
#
# ============================================================
file_path = "/content/drive/MyDrive/AI_멀티_MultiAgent/Multi_Agent_2607/Agent_2607/Agent_data/[이슈리포트 2022-2호] 혁신성장 정책금융 동향.pdf"
loader = PyPDFLoader(file_path)

# ============================================================
# TODO 2
#
# 파일의 확장자를 추출하세요.
#
# Hint)
# os.path.splitext()
#
# ============================================================
extension = os.path.splitext(file_path)[1].lower()


# ============================================================
# TODO 3
#
# 확장자에 따라 적절한 Loader를 생성하세요.
#
# pdf  -> PyPDFLoader
# docx -> Docx2txtLoader
# csv  -> CSVLoader
#
# 지원하지 않는 파일은 오류를 발생시키세요.
#
# ============================================================
if extension == ".pdf":
  loader = PyPDFLoader(file_path)

elif extension == ".docx":
  loader = Docx2txtLoader(file_path)

elif extension == ".csv":
  loader = CSVLoader(file_path)

else:
  raise ValueError(f"지원하지 않는 파일 형식({extension})")

# ============================================================
# TODO 4
#
# 문서를 읽으세요.
#
# Hint)
# loader.load()
#
# ============================================================

documents = loader.load()

# ============================================================
# TODO 5
#
# 여러 Document의 page_content를 하나의 문자열(text)로 만드세요.
#
# Hint)
#
# "\n".join(...)
#
# ============================================================

text = "\n".join(doc.page_content for doc in documents)

# ============================================================
# TODO 6
#
# PromptTemplate을 작성하세요.
#
# 요구사항
#
# - 핵심 내용을 요약
# - 5줄 이내
#
# ============================================================

prompt = PromptTemplate (
    template= """
    당신은 문서 분석 전문가 입니다.
    다음 문서를 읽고 핵심 내용을 5줄 이내로 요약하세요.

    ========================
    문서 내용
    ========================
    {text}

    ========================
    출력 형식
    ========================
    - 핵심내용1
    - 핵심내용2
    - 핵심내용3

    """,
    input_variables= ["text"]
)


# ============================================================
# TODO 7
#
# ChatOpenAI 객체를 생성하세요.
#
# 모델
#
# gpt-4.1-mini
#
# ============================================================

llm = ChatOpenAI(model = "gpt-4o", temperature= 0)

# ============================================================
# TODO 8
#
# LCEL Chain을 작성하세요.
#
# Prompt
# ↓
# ChatOpenAI
#
# ============================================================

chain = prompt | llm

# ============================================================
# TODO 9
#
# invoke()를 이용하여 문서를 요약하세요.
#
# ============================================================

result = chain.invoke({"text" : text})

# ============================================================
# TODO 10
#
# 결과를 출력하세요.
#
# ============================================================

print("="*60)
print("문서 요약 결과")

print("="*60)
print(result.content)
```

문제) Multi Document Analyzer(여러 문서 업로드 및 통합 분석)

```python

# ============================================================
#
# [학습 목표]
# 1. 여러 개의 문서를 업로드하여 분석할 수 있다.
# 2. 파일 확장자에 따라 적절한 Document Loader를 선택할 수 있다.
# 3. 여러 Document를 하나의 문자열로 결합할 수 있다.
# 4. LangChain LCEL을 이용하여 여러 문서를 종합 분석할 수 있다.
#
# ============================================================
# 문제
# ============================================================
#
# 사용자가 PDF, Word(docx), CSV 문서를 여러 개 업로드하면
# 파일 확장자에 맞는 Document Loader를 자동으로 선택하여
# 모든 문서를 읽고 하나의 문자열로 결합한 후
# AI를 이용하여 종합 분석하는 프로그램을 작성하세요.
#
# ============================================================
# 요구사항
# ============================================================
#
# 1. 업로드한 파일 목록을 저장하세요.
# 2. 파일 확장자를 확인하세요.
# 3. 확장자에 맞는 Document Loader를 생성하세요.
# 4. 모든 문서를 읽으세요.
# 5. Document의 page_content를 하나의 문자열로 결합하세요.
# 6. PromptTemplate을 작성하세요.
# 7. ChatOpenAI 객체를 생성하세요.
# 8. LCEL Chain을 작성하세요.
# 9. 여러 문서를 종합 분석하세요.
# 10. 결과를 출력하세요.
#
# ============================================================
# 입력 예시
# ============================================================
#
# report.pdf
# meeting.docx
# sales.csv
#
# ============================================================
# 출력 예시
# ============================================================
#
# ==============================
# Multi Document Analyzer
# ==============================
#
# [문서별 핵심 내용]
# ...
#
# [공통 핵심 키워드]
# ...
#
# [전체 종합 요약]
# ...
#
# [향후 액션]
# ...
#
# ============================================================

import os

from langchain_community.document_loaders import (
    PyPDFLoader,
    Docx2txtLoader,
    CSVLoader,
)

from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

# ============================================================
# TODO 1
#
# 업로드한 파일 목록을 저장하세요.
# 로컬 PC에서 여러 개의 문서를 업로드한다.
# 업로드한 파일은 Dictionary 형태로 반환되며,
# 파일 이름만 추출하여 리스트로 저장한다.
#
# ============================================================
from google.colab import files

uploaded = files.upload()

files_path = list(uploaded.keys())

# ============================================================
# TODO 2
#
# 모든 Document를 저장할 리스트를 생성하세요.
#
# ============================================================

documents = []

# ============================================================
# TODO 3
#
# 업로드한 파일을 하나씩 반복하면서
# 파일 확장자를 추출하세요.
#
# Hint)
#
# os.path.splitext()
#
# ============================================================

for file_path in files_path:

    extension = os.path.splitext(file_path)[1].lower()

# ============================================================
# TODO 4
#
# 파일 확장자에 따라
# 적절한 Document Loader를 생성하세요.
#
# PDF  -> PyPDFLoader
# DOCX -> Docx2txtLoader
# CSV  -> CSVLoader
#
# 지원하지 않는 파일은 건너뛰세요.
#
# ============================================================
    if extension == ".pdf":
        loader = PyPDFLoader(file_path)
    elif extension == ".docx":
        loader = Docx2txtLoader(file_path)
    elif extension == ".csv":
        loader = CSVLoader(file_path)
    else :
        print(f"{file_path} : 지원하지 않는 파일입니다.")
        continue
# ============================================================
# TODO 5
#
# 문서를 읽어 documents 리스트에 저장하세요.
#
# Hint)
#
# loader.load()
# documents.extend(...)
#
# ============================================================
    documents.extend(loader.load())


# ============================================================
# TODO 6
#
# 모든 Document의 page_content를
# 하나의 문자열(text)로 결합하세요.
#
# Hint)
#
# "\n\n".join(...)
#
# ============================================================

text = "\n".join(doc.page_content for doc in documents)

# ============================================================
# TODO 7
#
# PromptTemplate을 작성하세요.
#
# 요구사항
#
# - 문서별 핵심 내용
# - 공통 키워드
# - 전체 종합 요약
# - 향후 액션
#
# ============================================================

prompt = PromptTemplate (
    template= """
    당신은 문서 분석 전문가 입니다.
    다음 문서를 읽고 핵심 내용을 5줄 이내로 요약하세요.
    업로드된 모든 파일의 내용을 요약 해야 한다.

    ========================
    문서 내용
    ========================
    {text}

    ========================
    출력 형식
    ========================
    - 핵심내용1
    - 핵심내용2
    - 핵심내용3

    """,
    input_variables= ["text"]
)


# ============================================================
# TODO 8
#
# ChatOpenAI 객체를 생성하세요.
#
# 모델
#
# gpt-4.1-mini
#
# ============================================================

llm = ChatOpenAI(model = "gpt-4o", temperature= 0)

# ============================================================
# TODO 9
#
# LCEL Chain을 작성하세요.
#
# Prompt
# ↓
# ChatOpenAI
#
# ============================================================

chain = prompt | llm

# ============================================================
# TODO 10
#
# invoke()를 이용하여
# 여러 문서를 종합 분석하세요.
#
# ============================================================

result = chain.invoke({"text" : text})

# ============================================================
# TODO 11
#
# 결과를 출력하세요.
#
# ============================================================
print("="*60)
print("문서 요약 결과")

print("="*60)
print(result.content)
```
